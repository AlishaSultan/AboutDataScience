# Categorial plot:

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
sns.countplot(x = 'smoker', data=df)
```

```python
sns.countplot(x = 'day', data=df)
```

```python
sns.countplot(y = 'Gender', data=df)
```

```python
sns.barplot(x = 'total_bill' , y = 'Gender' , data = df)
```

```python
sns.boxplot(y = 'total_bill', x = 'Gender' ,  data = df)
```

```python
sns.boxplot(y = 'total_bill', x = 'Gender' ,  data = df , palette = 'rainbow')
```

```python
sns.boxplot(y = 'total_bill', x = 'Gender' ,  data = df , hue = 'smoker')
```

```python
sns.violinplot(y = 'total_bill' , x = 'Gender' ,data = df, hue = 'smoker')
```
